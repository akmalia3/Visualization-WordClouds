### **Case Folding**

  Case folding merupakan proses dalam text preprocessing yang dilakukan untuk menyeragamkan karakter pada data.

```python
# Membuat kolom articles untuk bahan nlp processing

df['articles'] = df['title'].str.cat(" " + df['description'])
df.head()
```

```python
# mengubah kata menjadi lowercase
df['articles to lower'] = df['articles'].str.lower()
df.head(5)
```

```python
! pip install swifter
```

```python
import re
import swifter
def remove_number(text):
    return  re.sub(r"\d+", "", text)

df['remove number'] = df['articles to lower'].swifter.apply(remove_number)
df.head()
```

```python
# Remove Puntuation, menghapus tanda baca

import string

def remove_punctuation(text): # fungsi untuk menghapus tanda baca
  return text.translate(str.maketrans('', '', string.punctuation))

# apply fungsi pada column
df['articles remove puntuation'] = df['remove number'].apply(remove_punctuation)
df['articles remove puntuation'].head(10)
```

```python
# Import library

import nltk
nltk.download('punkt')

from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
```

```python
# Tokenize

def word_tokenize_fungsi(text):
  return word_tokenize(text)

df['articles tokenize'] = df['articles remove puntuation'].apply(word_tokenize_fungsi)
df['articles tokenize']
```

### **Stopwords Removal**

  Stopword removal adalah proses menghapus kata-kata tidak penting atau kata hubung yang tidak mempengaruhi isi konten.

```python
# download dataset stopwords
!gdown (link) -O stopword.xlsx
```

```python
# import library
from nltk.corpus import stopwords
```

```python
list_stopwords = stopwords.words('indonesian') # deskripsikan stopword bahasa indonesia

stopwords = pd.read_excel('stopword.xlsx')

def split_word(a):
  return a.split(' ')

text = stopwords['stopwords'].apply(split_word)
single_list = [word for sublist in text for word in sublist] # split column into single list

list_stopwords.extend(single_list) # menambah stopword dari excel ke list stopwords

# jika error maka run dulu code sebelumnya
```

```python
list_stopwords = set(list_stopwords) # mengubah format list stopword menjadi dictionary

# fungsi remove stopwords
def remove_stopwords(words):
  return [word for word in words if word not in list_stopwords]

# menerapkan fungsi remove stopwords pada data
df['articles remove stopwords'] = df['articles tokenize'].apply(remove_stopwords)
df['articles remove stopwords'].head(20)
```

### **Stemming**
  
  Stemming adalah proses mengubah kata menjadi akar kata atau menghilangkan imbuhan pada kata.


```python
# import library

from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
# membuat stemmer
import swifter
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# fungsi stemming
def stemmer_words(word_list):
  return [stemmer.stem(words) for words in word_list]

# menerapkan fungsi pada kolom
df['articles stemming'] = df['articles remove stopwords'].swifter.apply(stemmer_words)
df['articles stemming'].head(10)
```

### **Normalization**

  Normalisasi kata dilakukan untuk mengubah variasi kata yang memiliki arti yang sama menjadi bentuk yang seragam. Variasi kata dapat berupa perbedaan akhiran atau awalan. Metode normalisasi kata umumnya menggunakan aturan gramatikal dan kamus.

  https://github.com/insomniagung/kamus_kbba/tree/main

```python
import pandas as pd
```

```python
# download dataset untuk normalisasi
!gdown (link) -O kbba.xlsx
```

```python
# membaca kbba dari excel
kbba_dictionary = pd.read_excel('kbba.xlsx')

# variabel kosong untuk menampung word normalization
normalizad_word_dict = {}

for index, row in kbba_dictionary.iterrows():
    if row[0] not in normalizad_word_dict:
        normalizad_word_dict[row[0]] = row[1]

# fungsi normalisasi
def normalized_term(document):
    return [normalizad_word_dict[term] if term in normalizad_word_dict else term for term in document]

# apply function
df['articles normalization'] = df['articles stemming'].apply(normalized_term)
df['articles normalization'].head(10)
```

##**WordClouds**

- **Single Grams**

```python
# import library
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# fungsi untuk generate bigrams dari nested list
def ngrams(text):
    text = ' '.join([str(e) for e in text]) # ubah nested list menjadi string
    return ''.join(text).strip() # strip untuk menghapus spasi ekstra di awal atau akhir kalimat
```

```python
b = df['articles normalization'].swifter.apply(ngrams)
ngrams = ' '.join(b)
```

```python
wordcloud = WordCloud(width = 2000, height = 1334,
                      random_state=1, collocations=False, normalize_plurals=False,
                      collocation_threshold = 2).generate(ngrams) # generate wordcloud

# visualisasi
plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### **Bigrams & Trigrams**


```python
# import library
from nltk.util import ngrams
import swifter
```

```python
# fungsi untuk generate bigrams dari nested list
def get_ngrams(text, n=2):
    text = ' '.join([str(e) for e in text]) # ubah nested list menjadi string
    n_grams = ngrams(text.split(), n) # pisah tiap kata dan gabung sesuai ngram yang didefinisikan (n)
    returnVal = []

    try:
        for grams in n_grams:
            returnVal.append('_'.join(grams)) #menggabungkan kata yang telah dipisah dengan _
    except(RuntimeError):
        pass

    return ' '.join(returnVal).strip() # strip untuk menghapus spasi ekstra di awal atau akhir kalimat
```

```python
# membuat kolom baru dan apply fungsi get_grams
bigrams_data = df['articles normalization'].swifter.apply(get_ngrams, n=2)

# ubah type kolom menjadi string
bigrams = ''.join(bigrams_data)
```

```python
# membuat kolom trigrams dan apply fungsi
trigrams_data = df['articles normalization'].swifter.apply(get_ngrams, n=3)

 # ubah list menjadi str
trigrams = ' '.join(trigrams_data)
```

```python
# generate dan set wordcloud
wordcloud_bigrams = WordCloud(width = 2000, height = 1334,
                              random_state=1, background_color='black',colormap='Pastel1',
                              collocations=False, normalize_plurals=False,
                              collocation_threshold = 2).generate(bigrams)

# visualisasi dengan matplotlib
plt.figure()
plt.imshow(wordcloud_bigrams, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# generate dan set wordcloud
wordcloud_trigrams = WordCloud(width = 2000, height = 1334,
                              random_state=1, background_color='white',colormap='gist_heat_r',
                              collocations=False, normalize_plurals=False,
                              collocation_threshold = 2).generate(trigrams)

# visualisasi dengan matplotlib
plt.figure()
plt.imshow(wordcloud_trigrams, interpolation='bilinear')
plt.axis("off")
plt.show()
```
